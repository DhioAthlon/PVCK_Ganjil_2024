In this step we will read images from urls, and display them using openCV, please note the difference when reading image in RGB and BGR format. The default input color channels are in BGR format for openCV. RGB?

```python
# Create a list to store the URLs of several images

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://upload.wikimedia.org/wikipedia/commons/thumb/f/f7/Alexandre_Cabanel_-_Fallen_Angel.jpg/350px-Alexandre_Cabanel_-_Fallen_Angel.jpg"]

# read and display image
# loop on each image url, several images can be stored in the list

for url in urls:
  image = io.imread(url)                            #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  #convert color to RGB
  final_frame = cv.hconcat((image, image_2))      #concatenate image
  cv2_imshow(final_frame)                         #show image
  print('\n')
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

height, width, _ = image_2.shape

#create a horizontal line in the middle of the image

for x in range(height):
  y = int((x / height) * width) # Calculate the corresponding y-coordinate for the diagonal
  if 0 <= y < width:
    image_3[x, y] = [255, 255, 255] # Set pixel to white

# Draw the line from bottom-left to top-right

for x in range(height):
  y = int((1 - x / height) * width) # Calculate the corresponding y-coordinate for the reverse diagonal
  if 0 <= y < width:
    image_3[x, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
